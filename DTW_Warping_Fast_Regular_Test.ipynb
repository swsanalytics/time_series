## DTW Warping Paths Fast Vs. Regular 

### Library Imports

```python
from dtaidistance import dtw
from dtaidistance import dtw_visualisation as dtwvis
import numpy as np

from dtaidistance import dtw
```

### Short series to long series

#### DTW Fast

```python
#dtw.distance_fast

s1 = np.array([0., 0, 1, 2, 1, 0, 1, 0, 0, 2, 1, 0, 0])
s2 = np.array([0., 1, 3, 1,  0,0, 0, 0, 0])
distance, paths = dtw.warping_paths_fast(s1, s2)
print(distance)
print(paths)
#The matrix with all warping paths can be visualised as follows:

#x = np.arange(0, 20, .5)
#s1 = np.sin(x)
#s2 = np.sin(x - 1)
d, paths = dtw.warping_paths(s1, s2, window=25, psi=2)
best_path = dtw.best_path(paths)
dtwvis.plot_warpingpaths(s1, s2, paths, best_path)
```

#### Regular DTW

```python

s1 = np.array([0., 0, 1, 2, 1, 0, 1, 0, 0, 2, 1, 0, 0])
s2 = np.array([0., 1, 3, 1,  0,0, 0, 0, 0])
distance, paths = dtw.warping_paths(s1, s2)
print(distance)
print(paths)
#The matrix with all warping paths can be visualised as follows:

#x = np.arange(0, 20, .5)
#s1 = np.sin(x)
#s2 = np.sin(x - 1)
d, paths = dtw.warping_paths(s1, s2, window=25, psi=2)
best_path = dtw.best_path(paths)
dtwvis.plot_warpingpaths(s1, s2, paths, best_path)
```
