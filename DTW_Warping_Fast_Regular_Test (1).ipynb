## DTW Warping Paths Fast Vs. Regular 

### Library Imports

```python
from dtaidistance import dtw
from dtaidistance import dtw_visualisation as dtwvis
import numpy as np

from dtaidistance import dtw
```

### Short series to long series

#### DTW Fast

```python

s1 = np.array([0., 0, 1, 2, 1, 0, 1, 0, 0, 2, 1, 0, 0])
s2 = np.array([0., 1, 3, 1,  0,0, 0, 0, 0])

d, paths = dtw.warping_paths_fast(s1, s2, window=25, psi=2)
best_path = dtw.best_path(paths)
dtwvis.plot_warpingpaths(s1, s2, paths, best_path)
print(paths)
```

#### Regular DTW

```python

s1 = np.array([0., 0, 1, 2, 1, 0, 1, 0, 0, 2, 1, 0, 0])
s2 = np.array([0., 1, 3, 1,  0,0, 0, 0, 0])

d, paths = dtw.warping_paths(s1, s2, window=25, psi=2)
best_path = dtw.best_path(paths)
dtwvis.plot_warpingpaths(s1, s2, paths, best_path)
print(paths)
```

## Equal Length Series

### Fast DTW

```python
#dtw.distance_fast

s1 = np.array([1., 0, 1, 2, 1, 0, 1, 0, 0, 2, 1, 1, 2])
s2 = np.array([8., 4, 1, 7, 1, 3, 1, 2, 1, 1, 4, 4, 4])

d, paths = dtw.warping_paths_fast(s1, s2, window=25, psi=2)
best_path = dtw.best_path(paths)
dtwvis.plot_warpingpaths(s1, s2, paths, best_path)

print(paths)
```

#### Regular DTW 

```python
#dtw.distance_fast
import numpy as np
from dtaidistance import dtw
s1 = np.array([1., 0, 1, 2, 1, 0, 1, 0, 0, 2, 1, 1, 2])
s2 = np.array([8., 4, 1, 7, 1, 3, 1, 2, 1, 1, 4, 4, 4])


d, paths = dtw.warping_paths(s1, s2, window=25, psi=2)
best_path = dtw.best_path(paths)
dtwvis.plot_warpingpaths(s1, s2, paths, best_path)
```

### Longer / Simulated Series of Unequal Length

```python
from statsmodels.tsa.arima_process import arma_generate_sample

# Length 2000

np.random.seed(42)
max_lag = 30

n = int(2000)
burn = 2000

alphas = np.array([-1, .4, 1, .2])
betas = np.array([0])

ar = np.r_[1, -alphas]
ma = np.r_[1, betas]

arma11 = arma_generate_sample(ar=ar, ma=ma, nsample=n, burnin=burn)
series1 = arma11.cumsum()


#Length 1500

np.random.seed(1111)
max_lag = 30

n = int(1500)
burn = 2000

alphas = np.array([-1, .4, 1, .2])
betas = np.array([0])

ar = np.r_[1, -alphas]
ma = np.r_[1, betas]

arma11 = arma_generate_sample(ar=ar, ma=ma, nsample=n, burnin=burn)
series2 = arma11.cumsum()
```

### DTW Fast

```python
#dtw.distance_fast
import timeit
start_time = timeit.default_timer()

s1 = series1
s2 = series2

d, paths = dtw.warping_paths_fast(s1, s2, window=25, psi=2)
best_path = dtw.best_path(paths)
dtwvis.plot_warpingpaths(s1, s2, paths, best_path)

print(paths)

elapsed = timeit.default_timer() - start_time
print(elapsed)
```
